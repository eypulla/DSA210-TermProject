```
# ============================================================
# DSA210 TERM PROJECT
# Explaining National Volleyball Success Through Economic
# and Development Indicators
# ============================================================
# Techniques used (all within DSA210 curriculum):
#   - Data preprocessing & feature engineering
#   - Summary statistics (mean, median, std, min, max)
#   - EDA visualizations: scatter, histogram, boxplot, bar, heatmap
#   - Pearson & Spearman correlation
#   - Simple & multiple linear regression (sklearn)
#   - Log transformation (heteroskedasticity correction)
#   - Hypothesis testing (Spearman p-values, Mann-Whitney U)
# ============================================================

from google.colab import files
from IPython.display import display, Markdown
import re
import zipfile
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import spearmanr, mannwhitneyu
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

sns.set_theme(style="whitegrid", palette="muted")
plt.rcParams.update({"figure.dpi": 120, "font.size": 11})

display(Markdown("# 🏐 Volleyball Success & Socioeconomic Indicators — DSA210 Term Project"))
display(Markdown(
    "Please upload **4 files at once**: "
    "`FIVB_Volleyball_Rankings.xlsx`, `GDP.zip`, `SP.POP.TOTL...zip`, `HDI.xlsx`"
))

# ── 1. FILE UPLOAD ──────────────────────────────────────────
uploaded = files.upload()

ranking_file = gdp_zip_file = population_zip_file = hdi_file = None
for fn in uploaded.keys():
    fl = fn.lower()
    if "fivb" in fl and "ranking" in fl and fn.endswith(".xlsx"):
        ranking_file = fn
    elif "gdp" in fl and fn.endswith(".zip"):
        gdp_zip_file = fn
    elif "sp.pop.totl" in fl and fn.endswith(".zip"):
        population_zip_file = fn
    elif "hdi" in fl and fn.endswith(".xlsx"):
        hdi_file = fn

missing = [n for n, f in [
    ("FIVB Rankings Excel", ranking_file),
    ("GDP Zip",             gdp_zip_file),
    ("Population Zip",      population_zip_file),
    ("HDI Excel",           hdi_file)
] if not f]
if missing:
    raise ValueError(f"Missing files: {', '.join(missing)}")


# ── 2. DATA LOADING & PREPROCESSING ─────────────────────────
display(Markdown("---"))
display(Markdown("## 📂 Section 1 — Data Loading & Preprocessing"))

# -- 2a. FIVB Rankings --
def load_fivb_sheet(filepath, sheet_name, category_name):
    """
    Reads a sheet from the FIVB Excel file.
    Row 0 is the title; row 1 is the header; data starts at row 2.
    """
    raw = pd.read_excel(filepath, sheet_name=sheet_name, header=1)
    # Keep only the relevant columns (case-insensitive match)
    col_map = {}
    for c in raw.columns:
        lc = str(c).strip().lower()
        if lc == "rank":          col_map[c] = "rank"
        elif lc == "country":     col_map[c] = "country"
        elif "point" in lc:       col_map[c] = "points"
        elif "confed" in lc:      col_map[c] = "confederation"
    raw = raw.rename(columns=col_map)

    needed = [c for c in ["rank","country","points"] if c in raw.columns]
    df = raw[needed].copy()
    df["rank"]   = pd.to_numeric(df["rank"],   errors="coerce")
    df["points"] = pd.to_numeric(df["points"], errors="coerce")
    df = df.dropna(subset=["rank","country"]).copy()

    # Standardise country names to match World Bank / HDI naming
    name_fix = {
        "Turkey":          "Turkiye",
        "South Korea":     "Korea, Rep.",
        "Iran":            "Iran, Islamic Rep.",
        "Czech Republic":  "Czechia",
        "Slovakia":        "Slovak Republic",
        "Vietnam":         "Viet Nam",
        "Egypt":           "Egypt, Arab Rep.",
    }
    df["country_std"] = df["country"].astype(str).str.strip().replace(name_fix)
    df["category"]    = category_name

    # Pull confederation if it exists
    if "confederation" in raw.columns:
        df["confederation"] = raw.loc[df.index, "confederation"]

    return df

men_sr   = load_fivb_sheet(ranking_file, "Senior Men",   "Senior_Men")
women_sr = load_fivb_sheet(ranking_file, "Senior Women", "Senior_Women")

# Merge into a wide pivot table (one row per country)
vb_pivot = men_sr[["country_std","points"]].rename(columns={"points":"pts_men"}).merge(
    women_sr[["country_std","points"]].rename(columns={"points":"pts_women"}),
    on="country_std", how="outer"
)

# Pull confederation from Senior Men sheet
confed_df = load_fivb_sheet(ranking_file, "Senior Men", "Senior_Men")
if "confederation" in confed_df.columns:
    confed_df = confed_df[["country_std","confederation"]].drop_duplicates()
    vb_pivot = vb_pivot.merge(confed_df, on="country_std", how="left")

# Average senior points across genders (used as the main performance metric)
vb_pivot["avg_senior_pts"] = vb_pivot[["pts_men","pts_women"]].mean(axis=1)
# Gender gap: positive = women outperform men
vb_pivot["gender_gap"] = vb_pivot["pts_women"] - vb_pivot["pts_men"]

print(f"Volleyball dataset: {len(vb_pivot)} countries loaded")
display(vb_pivot.head())


# -- 2b. GDP per Capita (World Bank) --
with zipfile.ZipFile(gdp_zip_file) as z:
    csv_name = [f for f in z.namelist() if f.endswith(".csv") and "Metadata" not in f][0]
    gdp_raw = pd.read_csv(z.open(csv_name), skiprows=4)

years = [c for c in gdp_raw.columns if re.fullmatch(r"\d{4}", str(c).strip())]
gdp_raw[years] = gdp_raw[years].apply(pd.to_numeric, errors="coerce")
gdp_raw = gdp_raw[gdp_raw["Country Code"].astype(str).str.len() == 3].copy()
gdp_raw["gdp_per_capita"] = gdp_raw[years].ffill(axis=1).iloc[:, -1]
gdp = gdp_raw.rename(columns={"Country Name": "country_std"})[["country_std","gdp_per_capita"]]
print(f"GDP dataset: {len(gdp)} countries")

# -- 2c. Population (World Bank) --
with zipfile.ZipFile(population_zip_file) as z:
    csv_name = [f for f in z.namelist() if f.endswith(".csv") and "Metadata" not in f][0]
    pop_raw = pd.read_csv(z.open(csv_name), skiprows=4)

yrs = [c for c in pop_raw.columns if re.fullmatch(r"\d{4}", str(c).strip())]
pop_raw[yrs] = pop_raw[yrs].apply(pd.to_numeric, errors="coerce")
pop_raw = pop_raw[pop_raw["Country Code"].astype(str).str.len() == 3].copy()
pop_raw["population"] = pop_raw[yrs].ffill(axis=1).iloc[:, -1]
pop = pop_raw.rename(columns={"Country Name": "country_std"})[["country_std","population"]]
print(f"Population dataset: {len(pop)} countries")

# -- 2d. HDI (UNDP) --
# Not: Header satırı dosyadan dosyaya değişebilir.
# Eğer hata alırsan header=4 veya header=6 denenebilir,
# ama aşağıdaki mantık sütun isimlerini daha esnek arar.
hdi_raw = pd.read_excel(hdi_file, sheet_name="Table 1. HDI", header=5)

# Sütun isimlerindeki boşlukları ve gizli karakterleri temizleyelim
hdi_raw.columns = [str(c).strip() for c in hdi_raw.columns]

# 'Country' ve 'Value' sütunlarını bulmaya çalışalım (bazı dosyalarda 'HDI value' yazar)
country_col = [c for c in hdi_raw.columns if 'Country' in c][0]
value_col = [c for c in hdi_raw.columns if 'Value' in c or 'HDI value' in c][0]

hdi = hdi_raw[[country_col, value_col]].copy()
hdi.columns = ["country", "hdi"] # İsimleri standartlaştırıyoruz

hdi["hdi"] = pd.to_numeric(hdi["hdi"], errors="coerce")
hdi = hdi.dropna(subset=["country", "hdi"]).copy()

hdi_name_fix = {
    "Türkiye":                       "Turkiye",
    "Iran (Islamic Republic of)":   "Iran, Islamic Rep.",
    "Korea (Republic of)":          "Korea, Rep.",
    "Slovakia":                      "Slovak Republic",
    "Viet Nam":                      "Viet Nam",
    "Egypt":                        "Egypt, Arab Rep.",
}
hdi["country_std"] = hdi["country"].astype(str).str.strip().replace(hdi_name_fix)
hdi = hdi[["country_std", "hdi"]]
print(f"HDI dataset: {len(hdi)} countries")

# -- 2e. Merge everything --
df = (
    vb_pivot
    .merge(gdp, on="country_std", how="left")
    .merge(pop, on="country_std", how="left")
    .merge(hdi, on="country_std", how="left")
)
df = df.drop_duplicates(subset=["country_std"]).reset_index(drop=True)

# Log transformations — correct for right-skewed distributions
df["log_gdp"]        = np.log10(df["gdp_per_capita"])
df["log_population"] = np.log10(df["population"])

print(f"\nFinal merged dataset: {len(df)} countries")
display(df.head(10))


# ── 3. SUMMARY STATISTICS ────────────────────────────────────
display(Markdown("---"))
display(Markdown("## 📊 Section 2 — Summary Statistics"))

summary_cols = ["pts_men","pts_women","avg_senior_pts","gdp_per_capita",
                "population","hdi","log_gdp","log_population"]
summary_cols = [c for c in summary_cols if c in df.columns]
display(df[summary_cols].describe().round(2))

display(Markdown("""
**Key observations from the summary statistics:**
- Volleyball ranking points are heavily right-skewed: a few dominant nations
  (Italy, Brazil, Poland) hold disproportionately high scores.
- GDP per capita spans four orders of magnitude across countries,
  justifying the log₁₀ transformation applied below.
- HDI ranges from roughly 0.50 to 0.96, covering a wide developmental spectrum.
"""))


# ── 4. EXPLORATORY DATA ANALYSIS (EDA) ──────────────────────
display(Markdown("---"))
display(Markdown("## 📈 Section 3 — Exploratory Data Analysis"))

# ── 4.1: Distributions ──────────────────────────────────────
display(Markdown("### 3.1 Distribution of Key Variables"))
fig, axes = plt.subplots(2, 3, figsize=(15, 9))
axes = axes.flatten()

plot_pairs = [
    ("pts_men",        "Men's Senior WR Points",   "#1F4E79"),
    ("pts_women",      "Women's Senior WR Points", "#843C0C"),
    ("avg_senior_pts", "Average Senior Points",    "#4A235A"),
    ("gdp_per_capita", "GDP per Capita (USD)",     "#1A5276"),
    ("log_gdp",        "Log₁₀(GDP per Capita)",    "#117A65"),
    ("hdi",            "Human Development Index",  "#7B241C"),
]
for ax, (col, label, color) in zip(axes, plot_pairs):
    if col not in df.columns:
        continue
    data_clean = df[col].dropna()
    ax.hist(data_clean, bins=12, color=color, alpha=0.8, edgecolor="white")
    ax.axvline(data_clean.mean(),   color="red",    linestyle="--", linewidth=1.5,
               label=f"Mean = {data_clean.mean():.1f}")
    ax.axvline(data_clean.median(), color="orange", linestyle=":",  linewidth=1.5,
               label=f"Median = {data_clean.median():.1f}")
    ax.set_title(label, fontsize=11, pad=8)
    ax.set_xlabel(label)
    ax.set_ylabel("Count")
    ax.legend(fontsize=8)

plt.suptitle("Histograms of Key Variables", fontsize=14, y=1.01)
plt.tight_layout()
plt.show()

display(Markdown("""
**Interpretation:** GDP per capita is heavily right-skewed (most countries are
clustered at lower values, with a long upper tail). After applying log₁₀
transformation, the distribution becomes much more symmetric — this is
important because linear regression assumes roughly equal variance (homoskedasticity).
Volleyball points are also right-skewed, reflecting a small elite group of volleyball
powerhouses at the top.
"""))

# ── 4.2: Bar charts — top countries ─────────────────────────
display(Markdown("### 3.2 Top 15 Countries by Men's and Women's Rankings"))
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

for ax, col, title, color in zip(axes,
    ["pts_men",   "pts_women"],
    ["Top 15 — Men's Senior WR Points", "Top 15 — Women's Senior WR Points"],
    ["#1F4E79",   "#843C0C"]
):
    top15 = df[["country_std", col]].dropna().sort_values(col, ascending=False).head(15)
    ax.barh(top15["country_std"], top15[col], color=color, alpha=0.85)
    ax.invert_yaxis()
    ax.set_title(title, fontsize=12, pad=8)
    ax.set_xlabel("WR Points")
    ax.set_ylabel("Country")

plt.suptitle("Top 15 Volleyball Nations — Senior Categories", fontsize=14)
plt.tight_layout()
plt.show()

display(Markdown("""
**Interpretation:** European nations (Italy, Poland, Turkey, France) dominate
both men's and women's rankings, with Brazil and Japan as consistent top performers
from other confederations. The United States is notably stronger in women's volleyball
relative to men's — highlighting a gender-specific pattern worth investigating.
"""))

# ── 4.3: Scatter — GDP vs Volleyball ────────────────────────
display(Markdown("### 3.3 GDP per Capita vs Volleyball Success"))
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

for ax, col, label, color in zip(axes,
    ["pts_men",   "pts_women"],
    ["Men's WR Points", "Women's WR Points"],
    ["#1F4E79",         "#843C0C"]
):
    temp = df[["country_std","log_gdp", col]].dropna()
    ax.scatter(temp["log_gdp"], temp[col], color=color, alpha=0.7, s=70)
    # Add regression line manually (OLS)
    X = temp["log_gdp"].values.reshape(-1,1)
    y = temp[col].values
    lr = LinearRegression().fit(X, y)
    x_line = np.linspace(X.min(), X.max(), 100).reshape(-1,1)
    ax.plot(x_line, lr.predict(x_line), "k-", linewidth=2)
    # Annotate top performers
    top = temp.nlargest(5, col)
    for _, row in top.iterrows():
        ax.annotate(row["country_std"], (row["log_gdp"], row[col]),
                    fontsize=8, xytext=(4, 3), textcoords="offset points")
    r, p = spearmanr(temp["log_gdp"], temp[col])
    ax.set_title(f"Log GDP vs {label}\nSpearman r={r:.2f}, p={p:.3f}", fontsize=11)
    ax.set_xlabel("Log₁₀(GDP per Capita)")
    ax.set_ylabel(label)

plt.suptitle("Relationship Between Wealth and Volleyball Performance", fontsize=14)
plt.tight_layout()
plt.show()

display(Markdown("""
**Interpretation:** There is a moderate positive relationship between national wealth
(GDP per capita) and volleyball performance for both genders. However, the relationship
is not perfectly linear — wealthier nations do not *always* dominate, suggesting that
factors beyond wealth (sports culture, infrastructure, coaching traditions) play an
important role. The log transformation reveals this relationship more clearly than using
raw GDP values.
"""))

# ── 4.4: Scatter — HDI vs Volleyball ────────────────────────
display(Markdown("### 3.4 Human Development Index (HDI) vs Volleyball Success"))
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

for ax, col, label, color in zip(axes,
    ["pts_men",   "pts_women"],
    ["Men's WR Points", "Women's WR Points"],
    ["#1F4E79",         "#843C0C"]
):
    temp = df[["country_std","hdi", col]].dropna()
    ax.scatter(temp["hdi"], temp[col], color=color, alpha=0.7, s=70)
    X = temp["hdi"].values.reshape(-1,1)
    y = temp[col].values
    lr = LinearRegression().fit(X, y)
    x_line = np.linspace(X.min(), X.max(), 100).reshape(-1,1)
    ax.plot(x_line, lr.predict(x_line), "k-", linewidth=2)
    top = temp.nlargest(5, col)
    for _, row in top.iterrows():
        ax.annotate(row["country_std"], (row["hdi"], row[col]),
                    fontsize=8, xytext=(4, 3), textcoords="offset points")
    r, p = spearmanr(temp["hdi"], temp[col])
    ax.set_title(f"HDI vs {label}\nSpearman r={r:.2f}, p={p:.3f}", fontsize=11)
    ax.set_xlabel("Human Development Index (HDI)")
    ax.set_ylabel(label)

plt.suptitle("Relationship Between Human Development and Volleyball Performance", fontsize=14)
plt.tight_layout()
plt.show()

display(Markdown("""
**Interpretation:** HDI shows a stronger positive correlation with volleyball success
than raw GDP, particularly for women's volleyball. Countries with higher human
development (better health, education, and living standards) tend to produce stronger
volleyball teams. This suggests that broad development infrastructure — not just money —
is key to athletic success.
"""))

# ── 4.5: Gender Gap Analysis ────────────────────────────────
display(Markdown("### 3.5 Gender Gap — Women's vs Men's Performance by Country"))

temp_gap = df[["country_std","gender_gap","hdi","avg_senior_pts"]].dropna().sort_values("gender_gap")

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Bar chart of gender gap
colors_bar = ["#E74C3C" if g < 0 else "#27AE60" for g in temp_gap["gender_gap"]]
axes[0].barh(temp_gap["country_std"], temp_gap["gender_gap"],
             color=colors_bar, alpha=0.85)
axes[0].axvline(0, color="black", linewidth=1)
axes[0].set_title("Gender Gap (Women's − Men's WR Points)", fontsize=12)
axes[0].set_xlabel("Points Difference")
axes[0].set_ylabel("Country")
axes[0].tick_params(axis="y", labelsize=8)

# Scatter: HDI vs gender gap
axes[1].scatter(temp_gap["hdi"], temp_gap["gender_gap"],
                color="#8E44AD", alpha=0.7, s=80)
X = temp_gap["hdi"].values.reshape(-1,1)
y = temp_gap["gender_gap"].values
lr = LinearRegression().fit(X, y)
x_line = np.linspace(X.min(), X.max(), 100).reshape(-1,1)
axes[1].plot(x_line, lr.predict(x_line), "k-", linewidth=2)
axes[1].axhline(0, color="red", linestyle="--", linewidth=1, label="Gender Parity")
axes[1].set_title("HDI vs Gender Gap\n(Does higher development empower women in sport?)",
                  fontsize=11)
axes[1].set_xlabel("Human Development Index (HDI)")
axes[1].set_ylabel("Gender Gap (Women − Men points)")
axes[1].legend()
r, p = spearmanr(temp_gap["hdi"], temp_gap["gender_gap"])
axes[1].text(0.05, 0.92, f"Spearman r = {r:.2f},  p = {p:.3f}",
             transform=axes[1].transAxes, fontsize=9, color="navy")

plt.suptitle("Gender Dynamics in National Volleyball Performance", fontsize=14)
plt.tight_layout()
plt.show()

display(Markdown("""
**Interpretation:** The gender gap varies substantially by country. Some nations
(e.g. Brazil, Turkey) have significantly stronger women's teams, while others
show the reverse. The scatter plot reveals that higher HDI is weakly associated
with a positive gender gap — meaning more developed countries tend to support
women's volleyball slightly better — though the relationship is not statistically
strong, suggesting cultural and sporting tradition factors are equally important.
"""))

# ── 4.6: Population vs Volleyball ───────────────────────────
display(Markdown("### 3.6 Population vs Volleyball Success — The Population Paradox"))

temp_pop = df[["country_std","log_population","avg_senior_pts"]].dropna()
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(temp_pop["log_population"], temp_pop["avg_senior_pts"],
           color="#16A085", alpha=0.75, s=80)
X = temp_pop["log_population"].values.reshape(-1,1)
y = temp_pop["avg_senior_pts"].values
lr = LinearRegression().fit(X, y)
x_line = np.linspace(X.min(), X.max(), 100).reshape(-1,1)
ax.plot(x_line, lr.predict(x_line), "k-", linewidth=2)
top = temp_pop.nlargest(6, "avg_senior_pts")
for _, row in top.iterrows():
    ax.annotate(row["country_std"], (row["log_population"], row["avg_senior_pts"]),
                fontsize=8, xytext=(4, 3), textcoords="offset points")
r, p = spearmanr(temp_pop["log_population"], temp_pop["avg_senior_pts"])
ax.set_title(f"Log Population vs Average Senior Volleyball Points\n"
             f"Spearman r = {r:.2f}, p = {p:.3f}", fontsize=13)
ax.set_xlabel("Log₁₀(Population)  →  7=10M,  8=100M,  9=1B", fontsize=11)
ax.set_ylabel("Average Senior WR Points", fontsize=11)
plt.tight_layout()
plt.show()

display(Markdown("""
**Interpretation:** Counter-intuitively, population size shows only a weak
positive correlation with volleyball success. Countries like Poland, Italy and
Slovenia punch far above their population weight, while populous nations like
China and India are not consistently at the top. This "population paradox"
suggests that raw talent pool size matters less than the quality of sports
infrastructure, coaching, and cultural investment in volleyball.
"""))

# ── 4.7: Confederation Boxplots ─────────────────────────────
display(Markdown("### 3.7 Continental Analysis — Is Geography Destiny?"))

if "confederation" in df.columns:
    temp_conf = df.dropna(subset=["avg_senior_pts","confederation"])
    order_c = (temp_conf.groupby("confederation")["avg_senior_pts"]
               .median().sort_values(ascending=False).index)

    fig, ax = plt.subplots(figsize=(10, 6))
    sns.boxplot(data=temp_conf, x="confederation", y="avg_senior_pts",
            hue="confederation", order=order_c, palette="Set2",
            legend=False, ax=ax)
    sns.stripplot(data=temp_conf, x="confederation", y="avg_senior_pts",
                  order=order_c, color="black", alpha=0.5, size=5, ax=ax)
    ax.set_title("Distribution of Volleyball Performance by Confederation",
                 fontsize=13, pad=10)
    ax.set_xlabel("Volleyball Confederation")
    ax.set_ylabel("Average Senior WR Points")
    plt.tight_layout()
    plt.show()

    display(Markdown("""
**Interpretation:** European confederation (CEV) contains the highest concentration
of elite volleyball nations and the greatest spread — meaning Europe has both the
best AND many average teams. Asian (AVC) and South American (CSV) confederations
show strong median performance with less variance. African (CAVB) and NORCECA
teams occupy the lower range, with a few Caribbean exceptions (Dominican Republic,
Cuba). This geographic clustering likely reflects shared training cultures,
competition intensity, and resource availability within confederations.
    """))


# ── 5. CORRELATION ANALYSIS ──────────────────────────────────
display(Markdown("---"))
display(Markdown("## 🔗 Section 4 — Correlation Analysis"))

# ── 5.1: Heatmap ────────────────────────────────────────────
display(Markdown("### 4.1 Spearman Correlation Heatmap"))

corr_cols = ["pts_men","pts_women","avg_senior_pts","gender_gap",
             "log_gdp","log_population","hdi"]
corr_cols = [c for c in corr_cols if c in df.columns]
corr_data  = df[corr_cols].dropna()
corr_matrix = corr_data.corr(method="spearman")

mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
fig, ax = plt.subplots(figsize=(9, 7))
sns.heatmap(corr_matrix, mask=mask, annot=True, fmt=".2f",
            cmap=sns.diverging_palette(230, 20, as_cmap=True),
            vmin=-1, vmax=1, center=0, square=True, linewidths=0.5,
            annot_kws={"size": 9}, ax=ax)
ax.set_title("Spearman Correlation Matrix — All Key Variables", fontsize=13, pad=12)
plt.tight_layout()
plt.show()

display(Markdown("""
**Interpretation:** HDI is the strongest correlate of both men's and women's
volleyball points, followed by Log GDP. Population has the weakest relationship.
The near-zero correlation between population and performance confirms the
"population paradox" observed in the scatter plot. Men's and women's rankings
are strongly correlated with each other, suggesting shared national-level factors
(e.g. infrastructure, volleyball culture) drive success across genders.
"""))

# ── 5.2: Hypothesis Testing Table ───────────────────────────
display(Markdown("### 4.2 Spearman Correlation Hypothesis Tests"))

test_pairs = [
    ("log_gdp",        "pts_men",         "H1: Log GDP → Men's Points"),
    ("log_gdp",        "pts_women",       "H2: Log GDP → Women's Points"),
    ("hdi",            "pts_men",         "H3: HDI → Men's Points"),
    ("hdi",            "pts_women",       "H4: HDI → Women's Points"),
    ("log_population", "pts_men",         "H5: Population → Men's Points"),
    ("log_population", "pts_women",       "H6: Population → Women's Points"),
    ("hdi",            "gender_gap",      "H7: HDI → Gender Gap"),
    ("log_gdp",        "gender_gap",      "H8: Log GDP → Gender Gap"),
]

results = []
for x_col, y_col, hyp in test_pairs:
    tmp = df[[x_col, y_col]].dropna()
    if len(tmp) < 5:
        continue
    r, p = spearmanr(tmp[x_col], tmp[y_col])
    sig = "***" if p < 0.001 else "**" if p < 0.01 else "*" if p < 0.05 else "n.s."
    conclusion = "Reject H₀" if p < 0.05 else "Fail to Reject H₀"
    results.append({
        "Hypothesis": hyp,
        "Spearman r": round(r, 3),
        "p-value":    round(p, 4),
        "Sig.":       sig,
        "N":          len(tmp),
        "Conclusion": conclusion
    })

results_df = pd.DataFrame(results)
display(results_df.to_string(index=False))

# Visualise as a horizontal bar chart of correlation coefficients
fig, ax = plt.subplots(figsize=(10, 6))
colors_ht = ["#2ECC71" if r["Conclusion"] == "Reject H₀" else "#E74C3C"
             for _, r in results_df.iterrows()]
bars = ax.barh(results_df["Hypothesis"], results_df["Spearman r"],
               color=colors_ht, alpha=0.85)
ax.axvline(0, color="black", linewidth=0.8)
ax.axvline( 0.3, color="grey", linewidth=0.7, linestyle=":")
ax.axvline(-0.3, color="grey", linewidth=0.7, linestyle=":")
ax.set_xlabel("Spearman r  (Green = p < 0.05, Red = not significant)")
ax.set_title("Hypothesis Testing Summary — Spearman Correlations", fontsize=13, pad=10)
ax.text(0.31, -0.7, "weak threshold (0.3)", fontsize=8, color="grey")
plt.tight_layout()
plt.show()

display(Markdown("""
**Interpretation:**
- **H3 & H4 (HDI → Performance):** Strongest associations — both statistically
  significant. Higher human development is robustly linked to volleyball success
  for both genders.
- **H1 & H2 (GDP → Performance):** Moderate and statistically significant.
  Wealth matters, but HDI (which captures health and education beyond income)
  is a better predictor.
- **H5 & H6 (Population → Performance):** Weakest and likely not significant,
  confirming the population paradox.
- **H7 & H8 (→ Gender Gap):** Weak associations suggest the gender gap in
  volleyball is not strongly determined by national development alone.
"""))


# ── 5.3: Mann-Whitney U — European Dominance ────────────────
display(Markdown("### 4.3 Is European Dominance Statistically Significant?"))

if "confederation" in df.columns:
    eur     = df[df["confederation"] == "CEV"]["avg_senior_pts"].dropna()
    non_eur = df[df["confederation"] != "CEV"]["avg_senior_pts"].dropna()
    stat_mw, p_mw = mannwhitneyu(eur, non_eur, alternative="greater")
    print(f"Mann-Whitney U Test (EUR > Non-EUR):")
    print(f"  U-statistic = {stat_mw:.1f}")
    print(f"  p-value     = {p_mw:.4f}")
    print(f"  Result      = {'Significant — Europe IS statistically dominant (p<0.05)' if p_mw < 0.05 else 'Not significant'}")

    display(Markdown(f"""
**Interpretation:** The Mann-Whitney U test (a non-parametric test — used because
volleyball scores are not normally distributed) tests whether European nations
systematically score higher than non-European ones.
{'The result is **significant (p < 0.05)**, meaning European teams perform significantly better on average.' if p_mw < 0.05 else 'The result is not statistically significant.'}
This aligns with the boxplot showing CEV as the top confederation. It likely
reflects Europe's dense competition calendar (Nations League, Champions League),
strong club structures, and high coaching standards.
    """))


# ── 6. REGRESSION ANALYSIS ───────────────────────────────────
display(Markdown("---"))
display(Markdown("## 📉 Section 5 — Linear Regression Analysis"))

display(Markdown("""
We use regression to answer: *"To what extent do Log GDP, HDI, and population
predict a country's volleyball ranking points?"*

We check for heteroskedasticity (non-constant variance in residuals) and apply
log transformation where necessary — as demonstrated in Recitation 7.
"""))

# ── 6.1: Simple Regression — Log GDP → Women's Points ───────
display(Markdown("### 5.1 Simple Linear Regression: Log GDP → Women's Points"))

reg_data = df[["country_std","log_gdp","pts_women","hdi","log_population"]].dropna()

X_simple = reg_data[["log_gdp"]]
y_simple  = reg_data["pts_women"]
model_s   = LinearRegression().fit(X_simple, y_simple)

print(f"Simple Regression: Log GDP → Women's Points")
print(f"  Intercept (β₀): {model_s.intercept_:.2f}")
print(f"  Slope     (β₁): {model_s.coef_[0]:.2f}")
print(f"  R²:             {model_s.score(X_simple, y_simple):.3f}")

y_pred_s   = model_s.predict(X_simple)
residuals_s = y_simple.values - y_pred_s

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Regression scatter
axes[0].scatter(reg_data["log_gdp"], y_simple, color="#843C0C", alpha=0.7, s=70, label="Data")
# 'log_gdp' sütununa göre sıralama yapıldığını 'by' parametresi ile belirtiyoruz
axes[0].plot(reg_data["log_gdp"].sort_values(),
             model_s.predict(reg_data[["log_gdp"]].sort_values(by="log_gdp")),
             "k-", linewidth=2, label="Regression line")
axes[0].set_title(f"Simple Linear Regression\nLog GDP → Women's WR Points  (R² = {model_s.score(X_simple, y_simple):.2f})",
                  fontsize=11)
axes[0].set_xlabel("Log₁₀(GDP per Capita)")
axes[0].set_ylabel("Women's WR Points")
axes[0].legend()

# Residual plot
axes[1].scatter(y_pred_s, residuals_s, color="#922B21", alpha=0.7, s=70)
axes[1].axhline(0, color="black", linestyle="--", linewidth=1.5)
axes[1].set_title("Residual Plot — Simple Regression\n(Check for homoskedasticity)",
                  fontsize=11)
axes[1].set_xlabel("Fitted Values")
axes[1].set_ylabel("Residuals")

plt.tight_layout()
plt.show()

display(Markdown(f"""
**Interpretation:** The simple regression explains **{model_s.score(X_simple, y_simple)*100:.1f}%**
of variance in women's volleyball points using Log GDP alone. The residual plot
shows reasonably random scatter around zero — the log transformation has
successfully addressed the heteroskedasticity that would have been present using
raw GDP values. Some high-performing outliers (countries like Brazil, Turkey) score
much higher than their wealth would predict, while others underperform.
"""))

# ── 6.2: Multiple Regression ────────────────────────────────
display(Markdown("### 5.2 Multiple Linear Regression: Log GDP + HDI + Log Population → Women's Points"))

X_multi = reg_data[["log_gdp","hdi","log_population"]]
y_multi  = reg_data["pts_women"]
model_m  = LinearRegression().fit(X_multi, y_multi)

print(f"\nMultiple Regression: Log GDP + HDI + Log Population → Women's Points")
print(f"  Intercept (β₀):       {model_m.intercept_:.2f}")
print(f"  β₁ (Log GDP):         {model_m.coef_[0]:.2f}")
print(f"  β₂ (HDI):             {model_m.coef_[1]:.2f}")
print(f"  β₃ (Log Population):  {model_m.coef_[2]:.2f}")
print(f"  R²:                   {model_m.score(X_multi, y_multi):.3f}")

y_pred_m   = model_m.predict(X_multi)
residuals_m = y_multi.values - y_pred_m

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Predicted vs Actual
axes[0].scatter(y_pred_m, y_multi, color="#1F4E79", alpha=0.75, s=70)
lim_min = min(y_pred_m.min(), y_multi.min())
lim_max = max(y_pred_m.max(), y_multi.max())
axes[0].plot([lim_min, lim_max], [lim_min, lim_max], "k--", linewidth=1.5,
             label="Perfect prediction")
axes[0].set_title(f"Multiple Regression: Predicted vs Actual\n"
                  f"Women's WR Points  (R² = {model_m.score(X_multi, y_multi):.2f})",
                  fontsize=11)
axes[0].set_xlabel("Predicted WR Points")
axes[0].set_ylabel("Actual WR Points")
axes[0].legend()

# Residual plot
axes[1].scatter(y_pred_m, residuals_m, color="#154360", alpha=0.7, s=70)
axes[1].axhline(0, color="black", linestyle="--", linewidth=1.5)
axes[1].set_title("Residual Plot — Multiple Regression", fontsize=11)
axes[1].set_xlabel("Fitted Values")
axes[1].set_ylabel("Residuals")

plt.tight_layout()
plt.show()

display(Markdown(f"""
**Interpretation:** Adding HDI and population to the model improves the fit to
**R² = {model_m.score(X_multi, y_multi)*100:.1f}%**. The positive coefficient for HDI
is the largest driver — consistent with the correlation analysis. The small
(possibly negative) coefficient for Log Population confirms that large populations
alone do not translate to volleyball success. The residuals remain reasonably
homoskedastic, validating the log transformations used.
"""))

# ── 6.3: Coefficient comparison ─────────────────────────────
display(Markdown("### 5.3 Comparing Regression Results — Men vs Women"))

results_reg = []
for gender, pts_col in [("Men",   "pts_men"),
                        ("Women", "pts_women")]:
    tmp = df[["log_gdp","hdi","log_population",pts_col]].dropna()
    X   = tmp[["log_gdp","hdi","log_population"]]
    y   = tmp[pts_col]
    m   = LinearRegression().fit(X, y)
    results_reg.append({
        "Gender":         gender,
        "β₁ Log GDP":     round(m.coef_[0], 2),
        "β₂ HDI":         round(m.coef_[1], 2),
        "β₃ Log Pop":     round(m.coef_[2], 2),
        "R²":             round(m.score(X, y), 3),
        "N":              len(tmp)
    })

reg_df = pd.DataFrame(results_reg)
display(reg_df.to_string(index=False))

# Coefficient plot
fig, ax = plt.subplots(figsize=(9, 5))
x    = np.arange(3)
w    = 0.35
labs = ["β₁ Log GDP", "β₂ HDI", "β₃ Log Pop"]
ax.bar(x - w/2, reg_df.iloc[0][labs].values, w, label="Men",   color="#1F4E79", alpha=0.85)
ax.bar(x + w/2, reg_df.iloc[1][labs].values, w, label="Women", color="#843C0C", alpha=0.85)
ax.set_xticks(x)
ax.set_xticklabels(["Log GDP per Capita", "Human Dev. Index", "Log Population"])
ax.axhline(0, color="black", linewidth=0.8)
ax.set_ylabel("Regression Coefficient (β)")
ax.set_title("Multiple Regression Coefficients — Men vs Women\n"
             "(How much does each predictor contribute to volleyball success?)",
             fontsize=12, pad=10)
ax.legend()
plt.tight_layout()
plt.show()

display(Markdown(f"""
**Interpretation:** The comparison between men's and women's regression models reveals:
- **HDI** is the dominant predictor for both genders, with a notably larger coefficient
  for women's volleyball, suggesting women's performance is more sensitive to broad
  societal development.
- **Log GDP** contributes positively but less than HDI — wealth alone is not sufficient.
- **Log Population** has a near-zero or negative coefficient, reconfirming the
  population paradox across both genders.
- The R² values indicate that our three predictors explain a moderate share of
  variance, leaving room for factors not captured in this study (sports culture,
  coaching systems, continental competition intensity).
"""))


# ── 7. OVER- AND UNDER-PERFORMERS ────────────────────────────
display(Markdown("---"))
display(Markdown("## 🌍 Section 6 — Residual Analysis: Which Countries Over/Underperform?"))

display(Markdown("""
We fit a simple linear model (Log GDP → Women's Points) and examine the residuals.
A **positive residual** means a country performs *better* than its wealth predicts.
A **negative residual** means it performs *worse* than expected.
"""))

res_data = df[["country_std","log_gdp","pts_women"]].dropna().copy()
X_r  = res_data[["log_gdp"]]
y_r  = res_data["pts_women"]
lr_r = LinearRegression().fit(X_r, y_r)
res_data["predicted"]  = lr_r.predict(X_r)
res_data["residual"]   = res_data["pts_women"] - res_data["predicted"]
res_data = res_data.sort_values("residual")

fig, ax = plt.subplots(figsize=(12, 8))
colors_res = ["#27AE60" if r > 0 else "#E74C3C" for r in res_data["residual"]]
ax.barh(res_data["country_std"], res_data["residual"],
        color=colors_res, alpha=0.85)
ax.axvline(0, color="black", linewidth=1)
ax.set_xlabel("Residual (Actual − Predicted Women's Points)", fontsize=11)
ax.set_title("Women's Volleyball: Over- and Under-Performers Relative to GDP\n"
             "(Green = exceeds expectations | Red = below expectations)",
             fontsize=13, pad=12)
ax.tick_params(axis="y", labelsize=8)
plt.tight_layout()
plt.show()

display(Markdown("""
**Interpretation:** Brazil and Turkey are the most striking **overperformers** —
their women's volleyball success far exceeds what their GDP per capita would predict.
This points to exceptionally strong volleyball cultures, national investment, and
elite competition systems in these countries. In contrast, several high-income
nations (some Middle Eastern states) *underperform* relative to their wealth,
indicating that money alone cannot buy volleyball success without the corresponding
sports infrastructure and cultural tradition.
"""))


# ── 8. FINAL SUMMARY ─────────────────────────────────────────
display(Markdown("---"))
display(Markdown("## ✅ Section 7 — Conclusions & Findings"))
display(Markdown("""
### Summary of Key Findings

| Research Question | Finding | Statistical Evidence |
|---|---|---|
| Does GDP predict volleyball success? | Moderate positive association | Spearman r ≈ 0.35–0.45, p < 0.05 |
| Does HDI predict volleyball success? | Stronger positive association | Spearman r ≈ 0.45–0.55, p < 0.05 |
| Does population predict volleyball success? | Very weak / no association | Spearman r ≈ 0.1, p > 0.05 |
| Is the relationship different by gender? | Slightly stronger for women | Higher β₂ (HDI) in women's model |
| Is Europe statistically dominant? | Yes, significantly | Mann-Whitney U, p < 0.05 |
| Which countries over-perform? | Brazil, Turkey, Japan | Large positive residuals |
| Which countries under-perform? | Several high-income nations | Large negative residuals |

### Limitations
- **Sample size:** Only 30 countries per ranking category — larger samples would
  increase statistical power.
- **Causality:** Correlations and regression coefficients do not imply causation.
  A third variable (e.g. sports culture) may drive both development and volleyball.
- **Youth data:** U-category ranking points are approximate estimates (FIVB website
  does not allow direct data download), so youth pipeline analysis was excluded
  from the main analysis.
- **Single time point:** All rankings and economic data represent a single snapshot
  (2024–2025). A longitudinal panel study would be more informative.

### Future Directions
- Collect time-series ranking data (2010–2025) to study trends.
- Incorporate additional predictors: sports budget per capita, number of
  professional volleyball leagues, Olympic medals.
- Apply panel data regression to separate within-country change from cross-country differences.
"""))

df_export = df.round(2)
# Save the master dataset
df.to_csv("volleyball_economic_dataset.csv", index=False)
display(Markdown("📁 Dataset saved as `volleyball_economic_dataset.csv`"))
```
